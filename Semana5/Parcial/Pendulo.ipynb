```python
#Definición de ecuaciones diferenciales

def r_dotdot(r,theta, r_dot, theta_dot):
    return (k/m) * (l-r) + g*np.cos(theta) + r*(theta_dot**2)

def theta_dotdot(r,theta, r_dot, theta_dot):
    return (-g/r)*np.sin(theta) - (2/r)*(r_dot*theta_dot)
```
